```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

performance_7 = pd.read_excel('ver7/performance.xlsx')
performance_8 = pd.read_excel('ver8/performance.xlsx')
performance_7 = performance_7.iloc[:,1:]
performance_8 = performance_8.iloc[:,1:]
performance_7.insert(0, 'ver', 7)
performance_8.insert(0, 'ver', 8)
performance = pd.concat([performance_7, performance_8])
```

# Grouping by Fleet Type

```python
# type0: 4x 30kW cars | 4 fast charge points
# type1: 4x 40kW cars | 4 fast charge points
# type2: 4x 70kW cars | 4 fast charge points
# type3: 4x 30kW cars | 2 fast charge points
# type4: 4x 40kW cars | 2 fast charge points
# type5: 4x 70kW cars | 2 fast charge points
```

```python
performance.groupby(['ver','fleetType']).mean().iloc[:,1:].round(2)
```

```python
rapidCharges = performance.groupby(['fleetType']).mean().iloc[:,2:-5]
cost = performance.groupby(['fleetType']).mean().iloc[:,-5:]
rapidCharges.plot()
cost.plot()
```

# Grouping by driving behaviour

```python
performance.groupby(['ver','mpkw']).mean().iloc[:,2:].round(2)
```

```python
rapidCharges = performance.groupby(['mpkw']).mean().iloc[:,3:-5]
cost = performance.groupby(['mpkw']).mean().iloc[:,-5:]
rapidCharges.plot()
cost.plot()
```

# Grouping by schedule

```python
# shift0 - work shift with break
# shift1 - work shift
# shift2 - work shift 2
# shift3 - first shift same (original)
# shift4 - second shift same
# shfit5 - national express
# shift6 - taxi
# shift7 - meal delivery
# shift8 - goods delivery
```

```python
performance.groupby(['ver','schedule']).mean().iloc[:,1:].round(2)
```

```python
rapidCharges = performance_7.groupby(['schedule']).mean().iloc[:,2:-5]
cost = performance_7.groupby(['schedule']).mean().iloc[:,-5:]

fig, ax = plt.subplots(figsize=(16,7))
ax.plot(rapidCharges, 'o-')
ax.grid(b=True, which='both', color='0.65', linestyle='-')
ax.set_xlabel('Schedule ID', fontsize=18)
ax.set_title('Rapid Charges', fontsize=20)
ax.legend(['dumb','leavetime','batt','battneeded/leavetime','cost-sensitive'])

fig2, ax2 = plt.subplots(figsize=(16,7))
ax2.plot(cost, 'o-')
ax2.grid(b=True, which='both', color='0.65', linestyle='-')
ax2.set_xlabel('Schedule ID', fontsize=18)
ax2.set_title('Cost', fontsize=20)
ax2.legend(['dumb','leavetime','batt','battneeded/leavetime','cost-sensitive'])
```
