```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

performance = pd.read_excel('performance.xlsx')
performance = performance.iloc[:,1:]
```

# Grouping by Fleet Type

```python
# type0: 4x 30kW cars | 4 fast charge points
# type1: 4x 40kW cars | 4 fast charge points
# type2: 4x 70kW cars | 4 fast charge points
# type3: 4x 30kW cars | 2 fast charge points
# type4: 4x 40kW cars | 2 fast charge points
# type5: 4x 70kW cars | 2 fast charge points

rapidCharges = performance.groupby(['fleetType']).mean().iloc[:,1:-5]
cost = performance.groupby(['fleetType']).mean().iloc[:,-5:]

fig, ax = plt.subplots(figsize=(16,7))
ax.plot(rapidCharges, 'o-')
ax.grid(b=True, which='both', color='0.65', linestyle='-')
ax.set_xlabel('Schedule ID', fontsize=18)
ax.set_title('Rapid Charges', fontsize=20)
ax.legend(['dumb','leavetime','battneeded','battneeded/leavetime','cost-sensitive'])

fig2, ax2 = plt.subplots(figsize=(16,7))
ax2.plot(cost, 'o-')
ax2.grid(b=True, which='both', color='0.65', linestyle='-')
ax2.set_xlabel('Schedule ID', fontsize=18)
ax2.set_title('Cost', fontsize=20)
ax2.legend(['dumb','leavetime','battneeded','battneeded/leavetime','cost-sensitive'])
```

# Grouping by schedule

```python
# shift0 - work shift with break              good
# shift1 - work shift
# shift2 - work shift 2
# shift3 - first shift same (original)        good
# shift4 - second shift same
# shfit5 - national express
# shift6 - taxi
# shift7 - meal delivery
# shift8 - goods delivery                     good
# shift9 - long work shift                    good

rapidCharges = performance.groupby(['schedule']).mean().iloc[:-1,1:-5]
cost = performance.groupby(['schedule']).mean().iloc[:-1,-5:]

fig, ax = plt.subplots(figsize=(16,7))
ax.plot(rapidCharges, 'o-')
ax.grid(b=True, which='both', color='0.65', linestyle='-')
ax.set_xlabel('Schedule ID', fontsize=18)
ax.set_title('Rapid Charges', fontsize=20)
ax.legend(['dumb','leavetime','battneeded','battneeded/leavetime','cost-sensitive'])

fig2, ax2 = plt.subplots(figsize=(16,7))
ax2.plot(cost, 'o-')
ax2.grid(b=True, which='both', color='0.65', linestyle='-')
ax2.set_xlabel('Schedule ID', fontsize=18)
ax2.set_title('Cost', fontsize=20)
ax2.legend(['dumb','leavetime','battneeded','battneeded/leavetime','cost-sensitive'])
```
